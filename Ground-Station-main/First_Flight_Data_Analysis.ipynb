```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv ('/content/flight_data1.csv')
df
```

```python
df = df.dropna()
```

```python
print("___________________ All values at maximum Altitude________________________")
connected.loc[df['Altitude'] == connected['Altitude'].max()]
```

```python
connected['Lat_degree'] = ((connected['Latitude']/100).fillna(0).astype(int))+((connected['Latitude']%100)/60)
connected['Lon_degree'] = ((connected['Longitude']/100).fillna(0).astype(int))+((connected['Longitude']%100)/60)

connected
```

```python
m = folium.Map(location = [connected['Lat_degree'].iloc[0],connected['Lon_degree'].iloc[0]], zoom_start = 17)
folium.CircleMarker(location = [connected['Lat_degree'].iloc[0],connected['Lon_degree'].iloc[0]],radius=150, fill_color='red').add_to(m)
marker = folium.Marker(
    location = [connected['Lat_degree'].iloc[0],connected['Lon_degree'].iloc[0]],
    popup="<stong>Initial Postion</stong>")

final_marker = folium.Marker(
    location = [connected['Lat_degree'].iloc[-1],connected['Lon_degree'].iloc[-1]],
    popup="<stong>Final Postion Recorded</stong>")

marker.add_to(m)
final_marker.add_to(m)

m
```

```python
plt.rcParams["figure.figsize"] = (25,10)
connected['Voltage'].plot()
plt.title("Plot of Voltage", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Voltage', fontsize= 20, fontweight='bold')


```

```python
connected['Altitude'].plot(style = '.')
plt.title("Plot of Altitude(Above Sea Level)", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Altitude', fontsize= 20, fontweight='bold')
```
