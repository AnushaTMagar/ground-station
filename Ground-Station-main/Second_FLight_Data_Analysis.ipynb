```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv ('/content/flight_data2.csv')
print("_____________ Not Preprocessed Data _________________________")
df
```

```python
df = df.dropna()
```

```python
print("____________________Preprocessed Data______________________")
df
```

```python
connected = df.loc[df['Rflink'] == 1]
disconnected = df.loc[df['Rflink'] == 0]
connected = connected.reset_index()
disconnected = disconnected.reset_index()
```

```python
print("_________________________________ Connected Data __________________________")
connected
```

```python
print("_____________________ Disconnected Data _______________________________-")
disconnected
```

```python
print("________Connected Data Count_________")
connected.count()
```

```python
print("________Disconnected Data Count_________")
disconnected.count()
```

```python

print("_____________ Max Data __________________")
connected.max()
```

```python
print("_____________ Min Data________________-")
connected.min()
```

```python
print("___________________ All values at maximum Altitude________________________")
connected.loc[connected['Altitude'] == connected['Altitude'].max()]
```

```python
plt.rcParams["figure.figsize"] = (35,10)
connected['Voltage'].plot()
plt.title("Plot of Voltage", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Voltage', fontsize= 20, fontweight='bold')


```

```python
connected['Altitude'].plot(style = '.')
plt.title("Plot of Altitude(Above Sea Level)", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Altitude', fontsize= 20, fontweight='bold')
```

```python
connected['Acceleration_x'].plot(label = 'Acceleration_x')
connected['Acceleration_y'].plot(label = 'Acceleration_y')
connected['Acceleration_z'].plot(label = 'Acceleration_z')
plt.title("Plot of Acceleration", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Acceleration', fontsize= 20, fontweight='bold')
plt.legend()
```

```python
plt.rcParams["figure.figsize"] = (25,10)
connected['Gyro_x'].plot(label = 'Gyro_x')
connected['Gyro_y'].plot(label = 'Gyro_y')
connected['Gyro_z'].plot(label = 'Gyro_z')
plt.title("Plot of Gyro", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Gyro', fontsize= 20, fontweight='bold')
plt.legend()
```

```python
plt.rcParams["figure.figsize"] = (25,10)
connected['Pressure'].plot(style = '.')
plt.title("Plot of Pressure", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Pressure', fontsize= 20, fontweight='bold')
```

```python
plt.rcParams["figure.figsize"] = (25,10)
connected['Temperature'].plot(style = '.')
plt.title("Plot of Temperature", fontsize= 20, fontweight='bold')
plt.xlabel('No. of Data', fontsize= 20, fontweight='bold')
plt.ylabel('Temperature', fontsize= 20, fontweight='bold')
```
